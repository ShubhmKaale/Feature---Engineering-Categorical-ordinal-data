```python
import pandas as pd
import datetime
```

```python
# create a variable  with dates and from that extract the weekdays

df_date = datetime.datetime.today()
df_date
```

```python
# create a list of dates with 20 days difference from today 
# dl =[]
#for x in range(0,20):
 #   d = df_date - datetime.timedelta(days=x)
  #  dl.append(d)


date_list =[df_date - datetime.timedelta(days=x) for x in range(0,20)]
date_list
```

```python
# converting the list to dataframe
df = pd.DataFrame(date_list)
df.columns=['day']
df
```

```python
#extracting the week day name

df['week_day'] = df['day'].dt.day_name()
df
```

# Feature Engineering for ordinal number replacement

```python
weekday_mapping = {'Monday':1,
                  'Tuesday':2,
                   'Wednesday':3,
                  'Thursday':4,
                   'Friday':5,
                   'Saturday':6,
                   'Sunday':7                   
                  }

df['ordinal_day'] = df['week_day'].map(weekday_mapping)
df.head(5)
```
